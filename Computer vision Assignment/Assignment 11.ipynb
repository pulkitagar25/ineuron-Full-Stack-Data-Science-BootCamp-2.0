## 1. What do REGION PROPOSALS entail?


Ans=>

Region Proposals refer to a technique used in object detection, where the goal is to identify and localize the objects of interest in an image. Region proposals involve generating a set of potential object locations or regions in the image and then using these proposals to guide the object detection process. The idea behind region proposals is to reduce the search space for object detection, making the process more efficient and faster. Common methods for generating region proposals include selective search, sliding windows, and R-CNNs.

## 2. What do you mean by NON-MAXIMUM SUPPRESSION? (NMS)


Ans=>

Non-maximum Suppression (NMS) is a post-processing technique used in object detection to eliminate overlapping bounding boxes that correspond to the same object in the image. NMS works by keeping the bounding box with the highest confidence score and removing all other overlapping boxes with lower scores. The goal of NMS is to ensure that each object in the image is represented by a single bounding box, even if multiple boxes were generated for the same object due to overlapping or ambiguity in the object's shape.

## 3. What exactly is mAP?


Ans=>

mAP stands for Mean Average Precision, and it is a metric used to evaluate the performance of object detection algorithms. The mAP metric calculates the average precision of the algorithm across all objects in the dataset, taking into account both the precision and recall of the algorithm. Precision measures the fraction of correctly detected objects among all objects detected by the algorithm, while recall measures the fraction of correctly detected objects among all objects present in the dataset. The mAP metric provides a single score that summarizes the overall accuracy of an object detection algorithm.

## 4. What is a frames per second (FPS)?



Ans=>

Frames per Second (FPS) is a measure of the number of individual frames or images that are processed in a video or image sequence in one second. FPS is commonly used to evaluate the performance of computer vision algorithms and to compare the processing speed of different systems. A higher FPS value indicates that the algorithm is able to process more images in a shorter amount of time, making the system more efficient and faster.

## 5. What is an IOU (INTERSECTION OVER UNION)?


Ans=>

Intersection Over Union (IOU) is a metric used to evaluate the accuracy of object detection algorithms. IOU measures the overlap between the ground truth bounding box for an object and the predicted bounding box generated by the algorithm. The IOU metric is calculated as the ratio of the area of overlap between the two bounding boxes to the area of the union of the two bounding boxes. A high IOU value indicates that the predicted bounding box is accurate and well-aligned with the ground truth bounding box, while a low IOU value indicates that there is a significant misalignment between the two boxes.

## 6. Describe the PRECISION-RECALL CURVE (PR CURVE)



Ans=>

Precision-Recall Curve (PR Curve) is a graphical representation of the precision and recall of a machine learning model, typically used to evaluate object detection algorithms. The x-axis of the PR curve represents recall, while the y-axis represents precision. The curve is generated by varying the threshold used to classify an object detection as positive or negative, and plotting the precision and recall values for each threshold. The PR curve provides a visual representation of the trade-off between precision and recall, and can be used to determine the optimal threshold for a given object detection task.

## 7. What is the term "selective search"?



Ans=>

Selective Search is a technique for generating region proposals for object detection in images. The selective search algorithm works by first creating a hierarchical segmentation of the image, and then merging similar segments based on color, texture, and other features. The resulting segments are then used as potential object locations, or region proposals, for the object detection algorithm. Selective search is a fast and effective way to generate region proposals, and has been widely used in object detection algorithms such as R-CNNs.

## 8. Describe the R-CNN model's four components.



Ans=>

The R-CNN model consists of four components:

- Region Proposal Network (RPN): generates a set of region proposals for potential object locations in the image.
- Feature Extraction Network: extracts high-level features from the region proposals.
- Classification Network: classifies the region proposals as containing an object or not.
- Bounding Box Regression Network: refines the location and size of the bounding box for each object in the image.

## 9. What exactly is the Localization Module?



Ans=>

The Localization Module is a component of object detection algorithms that is responsible for refining the location of objects in the image. The localization module takes the region proposals generated by the region proposal network and refines the bounding box location and size for each object in the image. The goal of the localization module is to improve the accuracy of the object detection by reducing the error in the location and size of the bounding boxes.

## 10. What are the R-CNN DISADVANTAGES?

Ans=>

R-CNNs have several disadvantages, including:

- High computational cost: R-CNNs are computationally expensive due to the need to extract features for each region proposal, classify each region, and perform bounding box regression.
- Slow inference speed: The slow inference speed of R-CNNs makes them less suitable for real-time object detection applications.
- Limited scalability: R-CNNs are not easily scalable to large datasets or to real-world object detection problems with many classes and object variations.
- Limited ability to capture spatial relationships: R-CNNs do not have the ability to capture spatial relationships between objects, making them less effective for detecting objects in complex scenes.
