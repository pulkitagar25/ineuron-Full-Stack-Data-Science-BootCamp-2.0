## 1. What is the COVARIATE SHIFT Issue, and how does it affect you?

Ans=>

The covariate shift issue occurs when the distribution of the input features changes during the training process, which can lead to poor model performance. This happens because the model is trained on a different distribution of the inputs than what it will be exposed to during deployment.

## 2. What is the process of BATCH NORMALIZATION?

Ans=>

Batch normalization is a normalization technique used in deep learning to normalize the activations of a layer for each mini-batch during training. The process involves computing the mean and standard deviation of the activations for each mini-batch and normalizing the activations with these values. This helps to ensure that the activations are zero-mean and have unit variance, which can improve the stability and convergence of the training process.

## 3. Using our own terms and diagrams, explain LENET ARCHITECTURE.




Ans=>

LeNet architecture is a convolutional neural network (CNN) architecture that was first used for handwritten digit recognition. It is a simple architecture that consists of alternating convolutional and max-pooling layers followed by fully connected layers. The convolutional layers are used to extract local features from the input image, while the max-pooling layers are used to reduce the spatial resolution of the activations, allowing the network to be invariant to small translations of the input. The fully connected layers are used to make a final prediction based on the features extracted by the convolutional and max-pooling layers.



## 4. Using our own terms and diagrams, explain ALEXNET ARCHITECTURE.


Ans=>

AlexNet architecture is a more advanced CNN architecture that was first used for image classification. It consists of multiple convolutional and max-pooling layers, interspersed with rectified linear unit (ReLU) activation functions and dropout layers to prevent overfitting. The architecture also includes normalization layers, which help to stabilize the training process. One of the key innovations in AlexNet was the use of multiple parallel convolutional and max-pooling layers, which allowed for a much deeper network architecture and better performance on the ImageNet dataset.



## 5. Describe the vanishing gradient problem.



Ans=>

The vanishing gradient problem occurs when the gradients in the backpropagation process are very small, which can slow down or prevent the training process from converging. This can occur in deep networks, especially when using activation functions such as the sigmoid or tanh functions, which have a saturating nonlinearity that can result in very small gradients. This can be mitigated by using activation functions such as ReLU, which have a linear nonlinearity and do not saturate, resulting in much larger gradients during backpropagation.

## 6. What is NORMALIZATION OF LOCAL RESPONSE?



Ans=>

Normalization of Local Response refers to the normalization technique used in the AlexNet architecture to counteract the vanishing gradient problem. This technique was introduced by Alex Krizhevsky, Ilya Sutskever, and Geoffrey Hinton in their 2012 paper on AlexNet. The normalization of local response involves normalizing the activations of each unit within a local region of the network, instead of normalizing all activations globally. This helps to maintain the distribution of activations across the network, making it easier for the gradients to flow through the network.

## 7. In AlexNet, what WEIGHT REGULARIZATION was used?


Ans=>

In AlexNet, weight regularization was used in the form of dropout, which randomly drops out a certain percentage of neurons during training. This helps to prevent overfitting by reducing the dependence of the model on any one feature.

## 8. Using our own terms and diagrams, explain VGGNET ARCHITECTURE.




Ans=>

VGGNet is a deep convolutional neural network architecture that was introduced in 2014 by the Visual Geometry Group at the University of Oxford. It is known for its simple and uniform architecture, which consists of repeating blocks of convolutional and max-pooling layers with increasing number of filters.

## 9. Describe VGGNET CONFIGURATIONS.


Ans=>

VGGNet configurations refer to the different variants of the VGGNet architecture, with varying depths and widths. Some common configurations include VGG16 and VGG19, which differ in the number of convolutional layers and the number of filters in each layer.

## 10. What regularization methods are used in VGGNET to prevent overfitting?



Ans=>

To prevent overfitting in VGGNet, weight decay and dropout were used as regularization techniques. Weight decay helps to reduce the magnitude of the weights in the network, preventing overfitting by reducing the model's capacity to fit the training data. Dropout randomly drops out neurons during training, making the model less dependent on any one feature.
